```python
import pandas as pd
import random
import numpy as np
```

```python
ans = ['not at all', 'several days', 'more than half of the days', 'nearly everyday']

df = pd.read_csv('mental_health_patients.csv')
```

```python
df.head(20)
```

```python
column_list = []

for col_name in range(len(df.columns)):
    column_list.append( df.columns[col_name])
column_list
```

```python
questionCode = ['little_interest', 'down_depressed','sleep_trouble','feeling_tired', 'appetite', 'self_pity','poor_concentration',
                'sluggish','suicidal_thoughts']
```

# Making a dictionary of the old and new values 

```python
oldNewColumn = dict(zip(column_list,questionCode))       
```

```python

df_new = df.rename(columns=oldNewColumn, inplace=False)
```

```python
df_new.describe()
```

```python
# encoding the answers 
for column in questionCode:
    df_new[column] = df_new[column].replace(ans,[0,1,2,3])
```

```python
df_new.tail(20)
```

### Depression scale 
1. 0 -4 minimal depression
2. 5-9 mild depression
3. 10-14 moderate depression
4. 15-19 moderate severe depression
5. 20-27 severe depression

```python
depression_scale = ['minimal', 'mild' , 'moderate', 'moderate severe', 'severe']
```

```python
def classifyScore(number):
    if (number < 5):
        return 'minimal'
    elif (number < 10 ):
        return "mild"
    elif (number < 15 ):
        return "moderate"
    elif (number < 20 ):
        return "moderate severe"
    else:
        return "severe"
 
```

## counting the values of each row and writing it's total

```python
df_new['score'] = df_new.sum(axis=1)
```

```python
df_new.tail(10)
```

```python
df_new.head()
```

```python
df_new['score'].describe()
```

```python
df_new['status'] = df_new['score'].apply(lambda x: classifyScore(x))
```

```python
df_new['status'].describe()
```

## encode the status 

```python
df_new['status'] = df_new['status'].replace(depression_scale,[0,1,2,3,4])
```

```python
df_new['status']
```

```python
df_new.columns
```

# splitting our data to training and test

```python
from sklearn.model_selection import train_test_split
```

```python
df_X = df_new[questionCode]
df_Y = df_new['status']
```

```python
df_X
```

```python
df_Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_X, df_Y, test_size=0.2)
```

# Training and using models to learn

# Random Forest classifier 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
rfc = RandomForestClassifier(n_estimators=500, random_state=40 )
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)
```

```python
ac_score = ( accuracy_score(y_test,y_pred) ) * 100

print('the score is ' + str(ac_score))
```

# Logistic regression prediction

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(random_state=40, max_iter=700)
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
ac_score = ( accuracy_score(y_test,y_pred) ) * 100

print('the score is ' + str(ac_score))
```
