```python
import pandas as pd
import random
import numpy as np

ans = ['not at all', 'several days', 'more than half of the days', 'nearly everyday']

def select_ans():
    global ans
    
    #get the index of the value
    val = random.choice(ans)
    return val


dataDict = {}

for user in range(1,20000):
    ansList = []
    for reply in range(9):
        ansList.append(select_ans())
        
    dataDict.update({user: ansList})
        
    

df = pd.DataFrame(dataDict)

df = df.transpose()
```
